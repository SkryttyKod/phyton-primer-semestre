```python
### ANALISIS DE DATOS PANDA ###

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#### INICIANDO EL DATAFRAME ####

datos = pd.read_csv("assets/data/games.csv", index_col="Id")
datos.head(20)
#datos.describe()
```

```python
#### DIMENSION DEL DATASET ####
print(f'Cantidad filas y columnas del Dataset: {datos.shape}')
```

```python
#### IDENTIFICANDO LAS COLUMNAS CON DATOS NULOS ####
datos.isnull().any()
```

```python
##### IDENTIFICANDO LAS FILAS EN ESPECIFICO CON DAGTOS NULOS #####

filas_nulas = datos[datos.isnull().any(axis=1)]

print(filas_nulas)
```

```python
#### ELIMINANDO LOS DATOS NULOS DEL DATASET ####
data_sin_nulos = datos.dropna()
data_sin_nulos
```

```python
#### FUNCION PARA VERIFICAR SI EL GENERO RPG ESTA PRESENTE EN LA LISTA ####
def filtro_rpg(genres):
    return 'RPG' in genres

# Aplica el filtro utilizando apply
datosRPG = datos[datos['Genres'].apply(filtro_rpg)]

# Muestra el dataset solamente incluyendo la columna
nombresRPG = datosRPG[['Title', 'Rating']]
nombresRPG.head(10)
```

```python
#### INFORMACION DEL DATASET YA FILTRADO ####
nombresRPG.info()
```

```python
#### IDENTIFICANDO SI EXISTEN DATOS NULOS EN EL NUEVO DATASET ####
nombresRPG.isnull()
```

```python
#### MOSTRANDO LOS PRIMEROS 10 VIDEOJUEGOS RPG Y SU RATING DE POPULARIDAD ####
datasetRPG = nombresRPG.head(10)
datasetRPG['Rating'].plot(kind='bar')

```

```python
#### MOSTRANDO LOS PRIMEROS 10 VIDEOJUEGOS RPG Y SU RANTING DE POPULARIDAD CON TITULO ####
datasetRPG = nombresRPG.head(10)
datasetRPG.set_index('Title', inplace=True) # Establecer la columna "Title" como indice
datasetRPG['Rating'].plot(x=None, kind='bar') # Utilizar el x en el eje
```

```python
#### AGREGAR ETIQUETAS AL GRAFICO ####
datasetRPG = nombresRPG.head(10)
datasetRPG.set_index('Title', inplace=True) # Establecer la columna "Title" como indice

ax = datasetRPG['Rating'].plot(x=None, kind='bar')

for i, rating in enumerate(datasetRPG['Rating']):
    ax.annotate(str(rating), xy=(i, rating), ha = 'center', va = 'bottom')
```

```python
#### PROMEDIO RATING VIDEOJUEGOS RPG ####
ratings_RPG = datos[datos['Genres'].apply(lambda x: 'RPG' in x)]['Rating']
promedio_RPG = ratings_RPG.mean()

promedio_RPG
```

```python
#### PROMEDIO RATINGS VIDEOJUEGOS PLATAFORMA ####
ratings_Platform = datos[datos['Genres'].apply(lambda x: 'Platform' in x)]['Rating']
promedio_Platform = ratings_Platform.mean()

promedio_Platform
```

```python
#### PROMEDIO RATINGS VIDEOJUEGOS AVENTURA ####
ratings_Aventura = datos[datos['Genres'].apply(lambda x: 'Adventure' in x)]['Rating']
promedio_Aventura = ratings_Aventura.mean()

promedio_Aventura
```

```python
#### PROMEDIO RATINGS VIDEOJUEGOS INDIE ####
ratings_Indie = datos[datos['Genres'].apply(lambda x: 'Indie' in x)]['Rating']
promedio_Indie = ratings_Indie.mean()

promedio_Indie
```

```python
####  crear un nuevo DataFrame que contiene los promedios de dos géneros (RPG e Indie) ####
dataset_promedios = pd.DataFrame({'Género': ['RPG', 'Indie', 'Platform', 'Adventure'], 'Promedio': [promedio_RPG, promedio_Indie, promedio_Platform, promedio_Aventura]})
dataset_promedios
```

```python
#### crea un gráfico de pastel que muestra los promedios de rating para cada género como porcentajes del total.


# En este caso, son los promedios de rating para cada género.
plt.pie(dataset_promedios['Promedio'], 

        # El argumento labels especifica las etiquetas de las secciones del gráfico de pastel. 
        # En este caso, son los nombres de los géneros.
        labels=dataset_promedios['Género'], 

        # El argumento autopct permite especificar un formato de cadena que se utilizará para etiquetar las secciones 
        # con su valor numérico como un porcentaje. En este caso, '%1.1f%%' significa que los porcentajes se mostrarán 
        # con un punto decimal de precisión.
        autopct='%1.1f%%')

# La función plt.title establece el título del gráfico.
plt.title('Promedio de Rating por Género')

# La función plt.show muestra el gráfico.
plt.show()
```
