```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Carga el data set
datos = pd.read_csv("assets/data/games.csv", index_col="Id")

# Muestra las primeras 5 filas
datos.head(5)
```

```python
# Revisa si hay alguna columna con un elemento nulo
print(datos.isnull().any())

# Muestra las filas con datos nulos en una columna especifica
print()
filas_con_nulo_datos = datos[datos['Rating'].isnull()]
print(filas_con_nulo_datos.head(13))

# Saber cuantas filas tienen un valor nulo en la columna "Rating"
print()
num_filas_con_nulo_datos = filas_con_nulo_datos.shape[0]
print('Numero de filas que tienen valor nulo en la columna "Rating":', num_filas_con_nulo_datos)
```

```python
##### REEMPLAZANDO DATOS NULOS DE LA COLUMNA RATING POR CEROS #####
datos['Rating'] = datos['Rating'].fillna(0)

# Verificando
print(datos.isnull().any())
```

```python
rating_menor_mayor_datos = datos.sort_values('Rating')
rating_menor_mayor_datos.head(20)
```

```python
rating_menor_mayor_datos_grafico = rating_menor_mayor_datos.head(20)

rating_menor_mayor_datos_grafico.set_index('Title', inplace=True)

rating_menor_mayor_datos_grafico['Rating'].plot(kind='bar')
```

```python
datos.to_csv('assets/data/games_sin_rating_nulos.csv')
rating_menor_mayor_datos.to_csv('assets/data/games_menor_mayor_rating.csv')
```
